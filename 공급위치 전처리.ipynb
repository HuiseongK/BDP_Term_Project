```python
import pandas as pd
import numpy as np

df = pd.read_csv("pjdata.csv", parse_dates=["계약시작일"])
df
```

```python
seoul = df[df['공급위치'].str.contains("서울특별시")]
seoul
```

```python
seoul2 = seoul.iloc[:, [10,12,13,26,33,34,35]]
seoul2.sort_values(by="공급위치")
```

```python
list = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구'
        , '도봉구', '동대문구', '동작구', '마포구', 
        '서대문구', '서초구', '성동구', '성북구', '송파구', 
        '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

for i in list:
    seoul2['공급위치'][(seoul2['공급위치'].str.contains(i))] = i
    
print(seoul2)
```

```python
import datetime

seoul2["year"] = seoul2["계약시작일"].dt.year
seoul2["month"] = seoul2["계약시작일"].dt.month
seoul2["year_month"] = seoul2['계약시작일'].apply(lambda x:x.strftime('%Y%m'))
x = seoul2.drop(["year","month","계약시작일"], axis = 1)
x
```

```python
seoul3 = x.rename(columns = {"year_month":"계약시작일"})
```

```python
seoul3
```

```python
x = seoul3.groupby(["공급위치","계약시작일"]).sum()
```

```python
x.to_csv("공급위치.csv", encoding = 'utf-8-sig')
```
